## Data processing of C++ script results

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# opening the csv file
df = pd.read_csv("data.csv")

# this line is only to show the data frame content
# the ".head()" method is to show only the number of lines that we pass as parameters
df.head(10)
```

```python
# calculating the mean of time for each n value and algorithm
df = df.groupby(['algorithm', 'n'])['time'].mean().reset_index()

df.head(10)
```

```python
# this is a way to make a new dataframe specifically for an algorithm

df_quick = df[df['algorithm'] == 'quick_sort']
df_quick.head(10)
```

```python
# the following line makes a dataframe out of a smaller interval of n values
df600 = df[df['n'] <= 600]

df600
```

```python
# the first line would be the same as:
# df_bubble = df[df['algorithm'] == 'bubble_sort']
# plt.plot(df_bubble['n'], df_bubble['time'], label='bubble sort')

# this cell creates a line chart for the bubble sort algorithm results
plt.plot(df[df['algorithm'] == 'bubble_sort']['n'], df[df['algorithm'] == 'bubble_sort']['time'], label='Bubble Sort')

plt.title('Bubble Sort')
plt.xlabel('n (in millions)')
# even with scientific notation i wanted to make it explicit
plt.ylabel('time (in seconds)')
plt.show()
```

```python
# this cell creates a line chart for the selection sort algorithm results
plt.plot(df[df['algorithm'] == 'selection_sort']['n'], df[df['algorithm'] == 'selection_sort']['time'], label='Selection Sort')

plt.title('Selection Sort')
plt.xlabel('n (in millions)')
plt.ylabel('time (in seconds)')
plt.show()
```

```python
# this cell creates a line chart for the insertion sort algorithm results
plt.plot(df[df['algorithm'] == 'insertion_sort']['n'], df[df['algorithm'] == 'insertion_sort']['time'], label='Insertion Sort')

plt.title('Insertion Sort')
plt.xlabel('n (in millions)')
plt.ylabel('time (in seconds)')
plt.show()
```

```python
# this cell creates a line chart for the shell sort algorithm results
plt.plot(df[df['algorithm'] == 'shell_sort']['n'], df[df['algorithm'] == 'shell_sort']['time'], label='shell sort')

plt.title('Shell Sort')
plt.xlabel('n (in millions)')
plt.ylabel('time (in seconds)')
plt.show()
```

```python
# this cell creates a line chart for the quick sort algorithm results
plt.plot(df[df['algorithm'] == 'quick_sort']['n'], df[df['algorithm'] == 'quick_sort']['time'], label='Quick Sort')

plt.title('Quick Sort')
plt.xlabel('n (in millions)')
plt.ylabel('time (in seconds)')
plt.show()
```

```python
# this cell creates a line chart for the merge sort algorithm results
plt.plot(df[df['algorithm'] == 'merge_sort']['n'], df[df['algorithm'] == 'merge_sort']['time'], label='Merge Sort')

plt.title('Merge Sort')
plt.xlabel('n (in millions)')
plt.ylabel('time (in seconds)')
plt.show()
```

```python
# this cell creates a line chart comparing the results of all six sorting algorithms
plt.plot(df[df['algorithm'] == 'bubble_sort']['n'], df[df['algorithm'] == 'bubble_sort']['time'], label='Bubble Sort')
plt.plot(df[df['algorithm'] == 'selection_sort']['n'], df[df['algorithm'] == 'selection_sort']['time'], label='Selection Sort')
plt.plot(df[df['algorithm'] == 'insertion_sort']['n'], df[df['algorithm'] == 'insertion_sort']['time'], label='Insertion Sort')
plt.plot(df[df['algorithm'] == 'shell_sort']['n'], df[df['algorithm'] == 'shell_sort']['time'], label='Shell Sort')
plt.plot(df[df['algorithm'] == 'quick_sort']['n'], df[df['algorithm'] == 'quick_sort']['time'], label='Quick Sort')
plt.plot(df[df['algorithm'] == 'merge_sort']['n'], df[df['algorithm'] == 'merge_sort']['time'], label='Merge Sort')

plt.xlabel('n (in millions)')
plt.ylabel('time (in seconds)')
plt.legend()
plt.show()
```

```python
# this cell creates a line chart comparing the results of O(n log n) sorting algorithms
plt.plot(df[df['algorithm'] == 'shell_sort']['n'], df[df['algorithm'] == 'shell_sort']['time'], label='Shell Sort')
plt.plot(df[df['algorithm'] == 'quick_sort']['n'], df[df['algorithm'] == 'quick_sort']['time'], label='Quick Sort')
plt.plot(df[df['algorithm'] == 'merge_sort']['n'], df[df['algorithm'] == 'merge_sort']['time'], label='Merge Sort')

plt.xlabel('n (in millions)')
plt.ylabel('time (in seconds)')
plt.legend()
plt.show()
```

```python
# this is a line chart for big-o notation
# code adapted from https://gist.github.com/niftycode/917bb33be4e91a10b1fffd020f987f94

# runtime comparisons set up
n = np.linspace(1, 10, 1000)
labels = ['O(1)', 'O(log n)', 'O(n)', 'O(n log n)', 'O(n)', 'O(n)', 'O(2^n)']
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, n**3, 2**n]

# plot setup
plt.ylim(0, 100)

for i in range(len(big_o)):
    plt.plot(n, big_o[i], label=labels[i])

plt.legend()
plt.ylabel('relative runtime')
plt.xlabel('n')
plt.show()
```

```python
# this is a cell to plot a table containing values of n from 10 to 1000000
# code adapted from https://towardsdatascience.com/simple-little-tables-with-matplotlib-9780ef5d0bc4
bS = []
sS = []
iS = []
shS = []
qS = []
mS = []

n_values = [10, 100, 1000, 10000, 100000, 1000000]

for n in n_values:
    bS_value = df[(df['algorithm'] == 'bubble_sort') & (df['n'] == n)]['time'].values[0]
    bS.append(bS_value)

    sS_value = df[(df['algorithm'] == 'selection_sort') & (df['n'] == n)]['time'].values[0]
    sS.append(sS_value)

    iS_value = df[(df['algorithm'] == 'insertion_sort') & (df['n'] == n)]['time'].values[0]
    iS.append(iS_value)

    shS_value = df[(df['algorithm'] == 'shell_sort') & (df['n'] == n)]['time'].values[0]
    shS.append(shS_value)

    qS_value = df[(df['algorithm'] == 'quick_sort') & (df['n'] == n)]['time'].values[0]
    qS.append(qS_value)

    mS_value = df[(df['algorithm'] == 'merge_sort') & (df['n'] == n)]['time'].values[0]
    mS.append(mS_value)

data =  [
    ['n = 10', 'n = 100', 'n = 1000', 'n = 10000', 'n = 100000', 'n = 1000000'],
    ['Bubble Sort'] + bS,
    ['Selection Sort'] + sS,
    ['Insertion Sort'] + iS,
    ['Shell Sort'] + shS,
    ['Quick Sort'] + qS,
    ['Merge Sort'] + mS
]

column_headers = data.pop(0)
row_headers = [x.pop(0) for x in data]

# formatting to 8 decimal places
cell_text = []
for row in data:
    cell_text.append([f'{x:1.8f}' for x in row])

rcolors = plt.cm.BuPu(np.full(len(row_headers), 0.05))
ccolors = plt.cm.BuPu(np.full(len(column_headers), 0.05))

plt.figure(linewidth=2, tight_layout={'pad':1})

the_table = plt.table(cellText=cell_text,
                      rowLabels=row_headers,
                      rowColours=rcolors,
                      rowLoc='right',
                      colColours=ccolors,
                      colLabels=column_headers,
                      loc='center')

the_table.scale(1, 1.5)
ax = plt.gca()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.box(on=None)

plt.show()
```
